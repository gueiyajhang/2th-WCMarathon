### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
res = BeautifulSoup(r.text, 'html.parser')
res        
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
<!--part_list_2 文字列表 開始-->
<div class="part_list_2">
<h3><span class="date">2020/04/13 21:51</span><em class="tag c_local">地方</em><a href="/news/20200413/1690695.htm" target="_blank">防疫越來越嚴　桃園洽公實名制4月20日起須戴口罩</a></h3>
'''
#方法一:
data=res.find('div',class_='part_list_2').find_all('h3')

for d in data:
    newstime=d.find(class_='date') #注意此處 .find()和 .find_all()差別
    newssort=d.find('em')
    newstitle=d.find('a')
    print(newstime.text,'[{}]'.format(newssort.string),newstitle.text)
    #注意 .string和 .text差別，參考https://zhuanlan.zhihu.com/p/30911642 
```

```python
#方法二:
import pandas as pd 
time=[]
sort=[]
title=[]
for d in data:
    newstime=d.find(class_='date') #注意此處 .find()和 .find_all()差別
    time.append(newstime.string)
    newssort=d.find('em')
    sort.append(newssort.string)
    newstitle=d.find('a')
    title.append(newstitle.text)

#建立dataframe
dataframe=pd.DataFrame({
    'Released time':time,
    'Sort':sort,
    'News title':title})
pd.set_option('display.max_rows', None) 
print(dataframe)
# 若結果被省略，可用pd.set_option('display.max_rows', None)顯示所有列數 和pd.set_option('display.max_columns', None)顯示所有欄數
```

```python
#方法三:
dataresult=pd.DataFrame(data,columns=['日期','類別','標題'])
dataresult 
#此處建立的dataframe是以原本解析後data直接建立，裡面含有標籤，會影響後續groupby等操作
#建立dataframe即使列表有標籤還是會只建立文字部分，小心後續groupby等操作，無法支持含有標籤的分類
```

### ③ 哪一個類別的文章最多

```python
datagroup=dataframe.groupby('Sort') 
# print(datagroup.size())#用.size()看每一組大小
table2=pd.DataFrame(datagroup.size(),columns=['文章數量'])
table2.sort_values(by = '文章數量',ascending=False)
```

```python
max_type = table2.sort_values(by=['文章數量'], ascending=False).index[0]
max_number = table2.sort_values(by=['文章數量'], ascending=False).values[0][0]
print('{}類別的文章最多, 有{}篇'.format(max_type, max_number))
```
